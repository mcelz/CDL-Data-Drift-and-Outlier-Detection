# Dimensionality Reduction for the Motion Sense Dataset 

```python
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.manifold import TSNE
from umap import UMAP
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import datasets
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import tensorflow
from scipy.stats import entropy
from scipy.special import rel_entr
from numpy.random import seed
from numpy.random import normal
```

```python
datao = pd.read_csv('datasets/revised_sub20_cleandata_forStream.csv')
```

```python
datao.shape
```

```python
data = datao.iloc[:,1:13]
```

```python
data.head(5)
```

# PCA

### PC = 2 

```python
n_components = 2
pca2 = PCA(n_components=n_components)
pca_fit2 = pca2.fit_transform(data)
df_pca2 = pd.DataFrame(data = pca_fit2, columns = ['PC '+ str(i+1) for i in range(n_components)])
```

```python
df_pca2 = pd.concat([df_pca2, datao[['outlier']]], axis = 1)
```

```python
df_pca2.head(5)
```

```python

plt.scatter(df_pca2[df_pca2['outlier']==0][['PC 1']], df_pca2[df_pca2['outlier']==0][['PC 2']], c='r', marker='^')
plt.scatter(df_pca2[df_pca2['outlier']==1][['PC 1']], df_pca2[df_pca2['outlier']==1][['PC 2']], c='b', marker='^')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Dimensional Reduction using PCA')
plt.show()
```

### PC = 3 

```python
n_components = 3
pca3 = PCA(n_components=n_components)
pca_fit3 = pca3.fit_transform(data)
df_pca3 = pd.DataFrame(data = pca_fit3, columns = ['PC '+ str(i+1) for i in range(n_components)])
```

```python
df_pca3 = pd.concat([df_pca3, datao[['outlier']]], axis = 1)
```

```python
x =df_pca3["PC 1"]
y =df_pca3["PC 2"]
z =df_pca3["PC 3"]
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=df_pca3["outlier"], marker='o', cmap = "BrBG" )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

Comment: Reducing dimensionality using PCA does not do a good job in separting outliers and normal data.

# TSNE

### n = 2 

```python
tSNE_model = TSNE(n_components=2)
fit = tSNE_model.fit_transform(data)
```

```python
df_tSNE = pd.DataFrame(data = fit , columns = ['PC '+ str(i+1) for i in range(2)])
df_tSNE = pd.concat([df_tSNE, datao[['outlier']]], axis = 1)
```

```python
plt.scatter(df_tSNE[df_tSNE['outlier']==0][['PC 1']], df_tSNE[df_tSNE['outlier']==0][['PC 2']], c='r', marker='^')
plt.scatter(df_tSNE[df_tSNE['outlier']==1][['PC 1']], df_tSNE[df_tSNE['outlier']==1][['PC 2']], c='b', marker='^')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Dimensional Reduction using PCA')
plt.show()
```

### n = 3 

```python
tSNE_model3 = TSNE(n_components=3)
fit3 = tSNE_model3.fit_transform(data)
```

```python
df_tSNE3 = pd.DataFrame(data = fit3 , columns = ['PC '+ str(i+1) for i in range(3)])
df_tSNE3 = pd.concat([df_tSNE3, datao[['outlier']]], axis = 1)
```

```python
x =df_tSNE3["PC 1"]
y =df_tSNE3["PC 2"]
z =df_tSNE3["PC 3"]
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=df_tSNE3["outlier"], marker='o', cmap = "BrBG" )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

Comment: Through the data visualizations, reducing dimensionality using tnse with n = 2 and n = 3 separated outliers from normal data. As is observed, outliers are in the margin of the data cluster.

# UMAP 

### n = 2

```python
umap_2d = UMAP(n_components=2, init='random', random_state=0)
proj_2d = umap_2d.fit_transform(data)
umap = pd.DataFrame(data = proj_2d , columns = ['PC '+ str(i+1) for i in range(2)])
df_umap = pd.concat([umap, datao[['outlier']]], axis = 1)
```

```python
plt.scatter(df_umap[df_umap['outlier']==0][['PC 1']], df_umap[df_umap['outlier']==0][['PC 2']], c='r', marker='^')
plt.scatter(df_umap[df_umap['outlier']==1][['PC 1']], df_umap[df_umap['outlier']==1][['PC 2']], c='b', marker='^')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Dimensional Reduction using UMAP')
plt.show()
```

### n = 3 

```python
umap_3d = UMAP(n_components=3, init='random', random_state=0)
proj_3d = umap_3d.fit_transform(data)
umap3 = pd.DataFrame(data = proj_3d , columns = ['PC '+ str(i+1) for i in range(3)])
df_umap3 = pd.concat([umap3, datao[['outlier']]], axis = 1)
```

```python
x =df_umap3["PC 1"]
y =df_umap3["PC 2"]
z =df_umap3["PC 3"]
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=df_umap3["outlier"], marker='o', cmap = "BrBG" )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

Comment: Dimensionality reduction using UMAP with n = 2 and n = 3 also did a good job in separating outliers and normal data. Outliers are gathered at the margin of the data cluster.
